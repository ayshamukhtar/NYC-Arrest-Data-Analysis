# Abstract

This project analyzes 2025 NYPD Arrest Data to uncover patterns in the severity of crimes across New York City. Our goal is to identify how the level of offense categorized as felony, misdemeanor, or violation, varies based on demographic and temporal features. We focus on three main research questions: (1) How does offense level relate to age group? (2) How does it vary across boroughs? (3) Are certain times like day of week or month more associated with serious offenses?

To address these questions, we conducted exploratory data analysis (EDA), cleaned and transformed the dataset, and engineered features including arrest month and weekday. We applied multiple classification models, Logistic Regression, Decision Tree, Random Forest, Poisson Regression, Support Vector Machine, and K-Nearest Neighbors and also implemented an ensemble voting classifier to improve predictive performance.

These insights can help inform patrol scheduling, youth intervention programs, and policy making for equitable public safety.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
url = "https://raw.githubusercontent.com/ayshamukhtar/CS381-Project-2-Proposal/main/NYPD_Arrest_Data__Year_to_Date__20250620.csv"
df = pd.read_csv(url)

# Display the first 5 rows
df.head()
```

## EDA

```python
df.describe()
```

Based on the above table, there are explicit missing values for the `KY_CD` column. We can also see the range of acceptable values for certain numeric columns: `ARREST_PRECINCT`, `JURISDICTION_CODE`, `PD_CD`, `KY_CD`.

```python
# Check for missing values
df.isna().sum().sort_values(ascending=False)
```

Based on the above table, a vast majority of the explicit missing values are contained within the `LAW_CAT_CD` column. The remainder of missing values are contained within the `KY_CD` column.

```python
df.info()
```

Based on the above table, each column has been assigned its proper data type. We can also confirm that there are certain columns that contain explicit missing values and that the dataset is composed of 71,242 observations.

```python
# Drop rows with any missing values
df.dropna(inplace=True)
```

We drop the rows that contain missing values since they only make up a very small percentage of the dataset.

```python
df['PERP_RACE'].value_counts()
```

Based on the table, the majority of perpetrators identify as either BLACK or WHITE HISPANIC. We also note that there is missing information regarding the race of 267 individuals that were arrested.

```python
df['LAW_CAT_CD'].value_counts()
```

Based on the above table, a majority of the offenses committed are classified as either misdemeanors or felonies. We also note that there are a few invalid values which include '9' and 'I'.

```python
# Drop rows with nonstandard values
df = df[df['LAW_CAT_CD'].isin(['F', 'M', 'V'])]

# View all unique values
print(df['LAW_CAT_CD'].value_counts())
```

We can drop the invalid values since our list of acceptable of values for LAW_CAT_CD only includes: felony (F), misdemeanor (M), and violation (V).

```python
# Convert 'ARREST_DATE' to datetime
df['ARREST_DATE'] = pd.to_datetime(df['ARREST_DATE'], errors='coerce')

# Extract time-based features
df['ARREST_MONTH'] = df['ARREST_DATE'].dt.month
df['ARREST_WEEKDAY'] = df['ARREST_DATE'].dt.day_name()
df['ARREST_YEAR'] = df['ARREST_DATE'].dt.year

# Preview the new columns
df[['ARREST_DATE', 'ARREST_MONTH', 'ARREST_WEEKDAY', 'ARREST_YEAR']].head()
```

To analyze trends in crime over time, we extracted new features from the `ARREST_DATE` column. These include the month of arrest, the day of the week, and the year. This allows us to group and visualize arrests by temporal patterns, which will be important for answering our third research question related to time-based crime trends.

#### Class Distribution

```python
df['LAW_CAT_CD'].value_counts().plot(kind='bar', title='Distribution of Offense Levels')
plt.ylabel('Count of Arrests')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Based on the plot, a majority of the offense levels recorded are misdemeanors and felonies.

#### Demographic Distributions by Offense Level

```python
df['AGE_GROUP'].value_counts().sort_index().plot(kind='bar', title='Arrests by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Arrests')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Based on the above plot, a majority of the arrests were made on young adults - middle aged adults that fall within the age group of 25-44.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='AGE_GROUP', hue='LAW_CAT_CD',
              order=df['AGE_GROUP'].value_counts().index,
              palette='pastel')
plt.title('Offense Level by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Arrests')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Offense Level by Age Group:

This bar chart shows the distribution of offense levels (LAW_CAT_CD) across different age groups. We observe that:
- The 25-44 age group is the most represented across all offense levels, especially for felony (F) and misdemeanor (M) arrests.
- Younger individuals (<18) and older individuals (65+) account for far fewer arrests, suggesting offenses are concentrated among middle-aged adults.
- Felonies and misdemeanors are the dominant categories in all age groups, with very few violations (V).
- **Insight for RQ1**: Most offenses, particularly felonies and misdemeanors, are associated with individuals aged 25-44.

#### Location and Time Distributions by Offense Level

```python
# Offense Level by Borough
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='ARREST_BORO', hue='LAW_CAT_CD',
              order=df['ARREST_BORO'].value_counts().index,
              palette='pastel')

plt.title('Offense Level by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Arrests')
plt.tight_layout()
plt.show()
```

Offense Level by Borough:

This plot breaks down arrests by borough and offense level:

```python
# Offense level by Weekday
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='ARREST_WEEKDAY', hue='LAW_CAT_CD',
              order=weekday_order, palette='pastel')

plt.title('Offense Level by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Number of Arrests')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Offense Level by Weekday:

This plot visualizes arrests by day of the week:
- Arrest levels are fairly consistent Monday through Friday, with a slight drop during weekends.
- Wednesday has the highest number of arrests, and again, felonies and misdemeanors are the most common.
- There is no dramatic weekday where offenses spike, suggesting steady police activity and incident rates throughout the week.
- **Insight for RQ3**: Offenses are slightly more frequent midweek (especially Wednesday), but levels remain fairly consistent across weekdays.

```python
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='ARREST_MONTH', hue='LAW_CAT_CD', palette='pastel')

plt.title('Offense Level by Month')
plt.xlabel('Month')
plt.ylabel('Number of Arrests')
plt.xticks(ticks=range(0, 12), labels=month_labels)
plt.tight_layout()
plt.show()
```
